```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rgb2gray(img):
    result = np.zeros([img.shape[0], img.shape[1]], dtype=np.uint8)
    R, G, B = cv2.split(img)
    result = 0.3*R + 0.59*G + 0.11*B
    return result
```

```python
img = cv2.imread("./inputs/h.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(img, cmap='gray')
plt.show()
```

```python
gray = rgb2gray(img)
plt.axis('off')
plt.imshow(gray, cmap='gray')
plt.show()
```
