Import libraries and images

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_1 = cv2.imread("./inputs/Hawkes_Bay_NZ.jpg", cv2.IMREAD_GRAYSCALE)
img_2 = cv2.imread("./inputs/4.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(img_1)
plt.show()
plt.imshow(img_2)
plt.show()
```

Display histograms

```python
hist_1 = cv2.calcHist([img_1], [0], None, [256], [0, 256])
hist_2 = cv2.calcHist([img_2], [0], None, [256], [0, 256])
plt.plot(hist_1)
plt.show()
plt.plot(hist_2)
plt.show()
```

improv images with histogram Equalization

```python
equ_1 = cv2.equalizeHist(img_1)
equ_2 = cv2.equalizeHist(img_2)
plt.imshow(np.hstack((img_1, equ_1)))
plt.show()
plt.imshow(np.hstack((img_2, equ_2)))
plt.show()
cv2.imwrite("./outputs/equ1.jpg", equ_1)
cv2.imwrite("./outputs/equ2.jpg", equ_2)
```

using clahe method

```python
img = cv2.imread('./inputs/tsukuba_l.png', 0)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl1 = clahe.apply(img)
cv2.imwrite('./outputs/clahe.jpg', cl1)
plt.imshow(cl1)
```
