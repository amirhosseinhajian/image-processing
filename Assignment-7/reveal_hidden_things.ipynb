Import libraries and image

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("./inputs/1.tif")
plt.imshow(img)
```

Define kernels

```python
kernels = [np.ones([5, 5])*0.04, np.ones([5, 5]), np.ones([5, 5])*5,\
           np.ones([3, 3])*0.04, np.ones([3, 3]), np.ones([3, 3])*5]
```

Make result

```python
copy_img = img
for kernel in kernels:
    result = cv2.filter2D(img, -1, kernel)
    copy_img = np.hstack((copy_img, result))
cv2.imwrite("./outputs/revealed.jpg", copy_img)
plt.imshow(copy_img)
```
