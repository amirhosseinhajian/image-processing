```python
from PIL import Image, ImageDraw, ImageFont, ImageOps
import matplotlib.pyplot as plt
import arabic_reshaper
from bidi.algorithm import get_display
```

Read image

```python
img = Image.open("./inputs/bg.jpg")
plt.axis('off')
plt.imshow(img)
plt.show()
```

Write persian text on image

```python
draw = ImageDraw.Draw(img)
text = get_display(arabic_reshaper.reshape(''))
draw.text((2250, 1100), text, font=ImageFont.truetype("./fonts/Yekan.ttf", 360), fill='rgb(255, 0, 255)')
plt.axis('off')
plt.imshow(img)
plt.show()
```

Calculate 3 histograms

```python
r, g, b = img.split()
plt.plot(r.histogram(), color='red')
plt.plot(g.histogram(), color='green')
plt.plot(b.histogram(), color='blue')
plt.show()
```

Equalize the image histogram

```python
equalized_img = ImageOps.equalize(img, mask=None)
r, g, b = equalized_img.split()
plt.plot(r.histogram(), color='red')
plt.plot(g.histogram(), color='green')
plt.plot(b.histogram(), color='blue')
plt.show()
plt.imshow(equalized_img, cmap='gray')
plt.show()
```

Convert image to gray

```python
gray = ImageOps.grayscale(img)
plt.axis('off')
plt.imshow(gray, cmap='gray')
plt.show()
```

Calculate histogram of gray image

```python
plt.plot(gray.histogram())
plt.show()
```

Equalize the gray image histogram

```python
equalized_gray = ImageOps.equalize(gray, mask=None)
plt.plot(equalized_gray.histogram())
plt.show()
plt.axis('off')
plt.imshow(equalized_gray, cmap='gray')
plt.show()
```
