```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("./inputs/spiderman.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
lower_hue = np.array([165, 95, 0]) 
upper_hue = np.array([180, 255, 255])
mask_red = cv2.inRange(img_hsv, lower_hue, upper_hue)
plt.imshow(mask_red, cmap='gray')
plt.axis('off')
plt.show()
```

```python
lower_hue = np.array([0, 115, 0]) 
upper_hue = np.array([15, 255, 255])
mask_red_2 = cv2.inRange(img_hsv, lower_hue, upper_hue)
plt.imshow(mask_red_2, cmap='gray')
plt.axis('off')
plt.show()
```

```python
result = img_hsv.copy()
result[:, : , 0] = result[:, : , 0].astype(np.float32)
mask = mask_red//255
result[:, :, 0] -= (mask * 145)
result[:, : , 0] = result[:, : , 0].astype(np.uint8)
mask = mask_red_2//255
result[:, :, 0] += (mask * 30)
result[:, : , 0] = result[:, : , 0].astype(np.uint8)
result = cv2.cvtColor(result, cv2.COLOR_HSV2RGB)
plt.imshow(result, cmap='gray')
plt.axis('off')
plt.show()
```

```python
lower_hue = np.array([100, 100, 0]) 
upper_hue = np.array([130, 255, 160])
mask_blue = cv2.inRange(img_hsv, lower_hue, upper_hue)
plt.imshow(mask_blue, cmap='gray')
plt.axis('off')
plt.show()
```

```python
mask_blue = mask_blue//255
result = cv2.cvtColor(result, cv2.COLOR_RGB2HSV)
result[:, : , 0] = result[:, : , 0].astype(np.float32)
result[:, :, 0] -= (mask_blue * 60)
result[:, : , 0] = result[:, : , 0].astype(np.uint8)
result = cv2.cvtColor(result, cv2.COLOR_HSV2RGB)
plt.imshow(result, cmap='gray')
plt.axis('off')
plt.show()
plt.imsave("./outputs/spiderman.jpg", result)
```
