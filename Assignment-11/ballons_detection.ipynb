```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("./inputs/balon.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
lower_hue = np.array([160,0,0]) 
upper_hue = np.array([180,255,255])
mask = cv2.inRange(img_hsv, lower_hue, upper_hue)//255
for i in range(3):
    img[:, :, i] = img[:, :, i]*mask 
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
plt.imsave("./outputs/ballon.png", img)
```
