```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./inputs/watermelon.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
h, s, v = cv2.split(img)
h = h.astype(np.float32)
green_indexes = np.where((h>30) & (h<80))
red_indexes1 = np.where(h<15)
red_indexes2 = np.where(h>165)
h[green_indexes] -= 50
h[h<0] += 180
h[red_indexes1] += 60
h[red_indexes2] += 60
result = cv2.merge([h.astype(np.uint8), s, v])
result = cv2.cvtColor(result, cv2.COLOR_HSV2RGB)
plt.imsave("./outputs/materwelon.jpg", result)
plt.imshow(result, cmap='gray')
plt.axis('off')
plt.show()
```
