```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("./inputs/44678.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
lower_hue = np.array([100, 80, 0]) 
upper_hue = np.array([140, 255, 255])
mask = cv2.inRange(img_hsv, lower_hue, upper_hue)
plt.imshow(mask, cmap='gray')
plt.axis('off')
plt.show()
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
result_green = img_hsv.copy()
result_green[:, : , 0] = result_green[:, : , 0].astype(np.float32)
mask = mask//255
result_green[:, :, 0] -= (mask * 60)
result_green[:, : , 0] = result_green[:, : , 0].astype(np.uint8)
result_green = cv2.cvtColor(result_green, cv2.COLOR_HSV2RGB)
plt.imshow(result_green, cmap='gray')
plt.axis('off')
plt.show()
plt.imsave("./outputs/cloth1.jpg", result_green)
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
result_orange = img_hsv.copy()
result_orange[:, : , 0] = result_orange[:, : , 0].astype(np.float32)
mask = mask//255
result_orange[:, :, 0] -= (mask * 100)
result_orange[:, : , 0] = result_orange[:, : , 0].astype(np.uint8)
result_orange = cv2.cvtColor(result_orange, cv2.COLOR_HSV2RGB)
plt.imshow(result_orange, cmap='gray')
plt.axis('off')
plt.show()
plt.imsave("./outputs/cloth2.jpg", result_orange)
```
